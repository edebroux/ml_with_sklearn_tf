### Performance measure(s)
We want to select a performance measure. For regression problems like this, it is more typical to use the root mean square error, RMSE, which is an estimate of the standard deviation (review MLE proof) $\hat{\sigma}$. How do we compute the RMSE?

$\begin{equation}
RMSE(\hat{y}, y) = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(\hat{y_i} - y_i)^2} = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(\mathbf{X_i}\hat{\beta} - y_i)^2}
\end{equation}$

It may be prudent to use the mean absolute error instead here too:

$\begin{equation}
MAE(\hat{y}, y) = \frac{1}{n} \sum_{i=1}^{n} | \hat{y_i} - y_i | = \frac{1}{n} \sum_{i=1}^{n} | \mathbf{X_i}\hat{\beta} - y_i |
\end{equation}$

The RMSE computes the *Euclidean norm*, i.e. the $\ell_2$ norm. The MAE calculates the $\ell_1$ norm. The $\ell_k$ norm is expressed as:

$\begin{equation}
|| \mathbf{v} ||_k = (|v_0|^k + \dots + |v_n|^k)^{1/k}
\end{equation}$

The $\ell_0$ norm gives the number the number of non-zero elements of the vector, and $\ell_{\infty}$ norm gives the maximum absolute value of the vector. Recall notes from MATH 350 @IIT.

### Create training and testing data sets
We want to be able to test to see if model is performing well, or if it is missing important patterns in the data that cause the estimates to be off. When you try to estimate the generalization error using the test set, the estimate may be to optimistic, and we will launch a system that will not perform as well as expected. This is called *data snooping bias*. We can create a test set quite easily through random selection.

````python
import numpy as np

def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]

# Define the randomly selected training and testing sets
train_set, test_set = split_train_test(housing, 0.20)
print("There are", len(train_set), "observations in the training set. \nThere are",
      len(test_set), "observations in the testing set.")
````

Although this works, the data generated will be different each time. If you set the random number generator's seed to the same value, such as `np.random.seed(42)`, before calling `np.random.permutation()`, you will get consistent sets.

These solutions will break the next time we try to use an updated dataset, so we could use each instance's indentifier to decide whether or not it should go in the test set. We could compute a hash of each instance's identifier and put the data in the test set if the hash is lower or equal to 20% of the maximum hash value. This ensures that the test set will remain consistnent across multiple runs, even if you refresh the dataset. Scikit-learn also provides a few functions to split datasets into multiple subsets in various ways. The simplest function is `train_test_split`, which does pretty much the same stuff as `split_train_test` shown above, except there are more parameters we can control, such as `random_state` and passing multiple datasets with an identical number of rows and the data will be split at the same indices.

````python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 0)
````

If the dataset is large enough, especially when considering the number of attributes, then a randomized train/test split is fine. However, smaller samples could run into sampling bias if random sampling is used. This is why there is a stratified sampling method written into the scikit-learn package.

Suppose we were told that median income is an important feature to predict housing prices. We want to make sure that the test set of data is representative of the various income. Looking at the the median income histogram, most of the incomes fall between `$15000` and `$60000`, but some median incomes go well beyond 6 (`$60000`). We should probably have not too many strata, and each stratum should be large enough. The following cut uses `pd.cut()` to create an income category attribute with 5 categories. Category 1 ranges from 0 to 1.5, cateogry 2 from 1.5 to 3.0, and so on. The final cateogry is all the median incomes that fall beyond 6.

````python
housing['income_category'] = pd.cut(housing['median_income'],
                                   bins = [0., 1.5, 3., 4.5, 6., np.inf],
                                   labels = [1, 2, 3, 4, 5])
````

Here is the histogramed `income_category` column:

Now you are ready to do stratified sampling based on the income category. For this, we can use `sklearn`'s `StratifiedShuffleSplit` class:

````python
from sklearn.model_selection import StratifiedShuffleSplit

# Random sampling to generate test set
train_set, test_set = train_test_split(housing, test_size = 0.2, random_state = 0)

# Stratified random sampling to generate test set
split = StratifiedShuffleSplit(n_splits = 1, test_size = 0.2, random_state = 0)
for train_index, test_index in split.split(housing, housing['income_category']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
````

Let's see if this stratification worked, starting by looking at the income category proportions in the test set:

### Custom Transformers
Scikit-Learn provides many useful transformers, sometimes we need to write our own for tasks such as custom cleanup operations or combining specific attributes. We will want our transformer to work seamlessly with Scikit-Learn functionalities, and since Scikit-Learn relies on duck typing (not inheritance, cool), all we need to create a class and implement three methods:
- `fit()`, which returns `self`
- `transform()`
- `fit_transform()`

We can get the last one for fee by simply adding the `TransformerMixin` as a base class. Also, if we add `BaseEstimator` as a base class, we will get two extra methods that will be useful for automatic hyperparamter tuning.

````python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_idx, bedrooms_idx, population_idx, households_idx = 3, 4, 5, 6

class CombineAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
        
    def fit(self, X, y = None):
        return self
    
    def transform(self, X, y = None):
        rooms_per_household = X[:, rooms_idx] / X[:, households_idx]
        population_per_household = X[:, population_idx] / X[: households_idx]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_idx] / X[:, rooms_idx]
            return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
````

We can add hyperparameters, such as `add_bedrooms_per_room`, to gate any data preparation step that we are not 100% certain about. The more you automate these data preparation steps, the more combinations you can automatically try out, making it much more likely that you will find a great combination.
### Feature Scaling
One of the most important transformations we need to apply to our data is *feature scaling*. With few exceptions, ML algorithms do not perform well when the input numerical attributes have very different scales. This is the casefor the housing data, so scaling is probably required. We could take one of two common approaches, *min-max scaling* or *standardization*.

**Transformation process for min-max scaling:**
Given a list of values $\{x_1, x_2, \dots, x_n\}$, min-max scaling converts a list of values to a 0-1 scale by using the transform:

$\begin{equation}
y_i = \frac{x_i - \ min \{x_1, x_2, \dots, x_n\}}{max \{x_1, x_2, \dots, x_n\} - \ min \{x_1, x_2, \dots, x_n\}}
\end{equation}$

where $i \in [n]$ and $y_i$ is the transformed value of $x_i \in \{x_1, x_2, \dots x_n\}$.

**Transformation process for standardization:**
Given a list of values $\{x_1, x_2, \dots, x_n\}$, standardization converts a list of values to a normally distributed variable with mean 0 and variance $\sigma^2$:

$\begin{equation}
y_i = \frac{x_i - \mu}{\sigma}
\end{equation}$, where $\mu$ is the mean of the list of values, and $\sigma$ is the standard deviation.



### Transformation Pipelines
We can create a transformation pipeline by using Scikit-Learn's `Pipeline` class. Here is an example of a small pipeline for numerical attributes:
```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, Imputer

num_pipeline = Pipeline([
    ('imputer', Imputer(strategy = "median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])
```
The `Pipeline` constructor takes a list of name/estimator pairs defining a sequence of steps. All but the last estimator must be transformers. The names can be anything we want. When we call the pipeline's `fit()` method, it calls `fit_transform()` sequentially on all transformers, passing the output of each call as the parameter to the next call, until it reachers the final estimator, for which it just calls the `fit()` method.

The pipeline exposes the same methods as the final estimator. In this example, the last estimator is a `StandardScaler`, which is a transformer, so the pipeline has a `transform()` method that applies all the transforms to the data in sequence.

So far, we have handled the categorical columns and the numerical columns separately. It would be more convenient to have a single transformer able to handle all columns, applying the appropriate transformations to each column.

````python
from sklearn.preprocessing import FunctionTransformer

rooms_ix, bedrooms_ix, population_ix, household_ix = [
    list(housing.columns).index(col)
    for col in ("total_rooms", "total_bedrooms", "population", "households")]

def add_extra_features(X, add_bedrooms_per_room=True):
    rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
    population_per_household = X[:, population_ix] / X[:, household_ix]
    if add_bedrooms_per_room:
        bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
        return np.c_[X, rooms_per_household, population_per_household,
                     bedrooms_per_room]
    else:
        return np.c_[X, rooms_per_household, population_per_household]

attr_adder = FunctionTransformer(add_extra_features, validate=False,
                                 kw_args={"add_bedrooms_per_room": False})
housing_extra_attribs = attr_adder.fit_transform(housing.values)
````

````python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', FunctionTransformer(add_extra_features, validate=False)),
        ('std_scaler', StandardScaler()),
    ])

housing_num_tr = num_pipeline.fit_transform(housing_tr)
housing_num_tr
````

In Scikit-Learn versions after `0.20` (after an update of my virtual environment, I have `0.24.1`), we can use `ColumnTransformer` for the purpose of dealing with Pandas DataFrame. Let's use it to apply all the transformations to the housing data:

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(housing_numeric)
cat_attribs = ['ocean_proximity']

# Create a pipeline that can deal with all of these attributes
full_pipeline = ColumnTransformer([
    ("numeric", num_pipeline, num_attribs),
    ("categorical", OneHotEncoder(), cat_attribs),
])

# Prepare the housing_data
housing_prepared = full_pipeline.fit_transform(housing)
````

Here is how this code works from `num_attribs` down:

- Lines 4 & 5: We get a list of numerical column names and categorical column names
- We start to construct a column transformer, lines 7 through 11
    - The constructor requires a list of tuples, where each tuple contains a name, a transformer, and a list of names or indices of columns that the transformer should be applied to
    - In this example, we specify that the numerical columns should be transformed using `num_pipeline` that we defined earlier
    - The categorical columns should be transformed using a `OneHotEncoder`
    - Finally, we apply this `ColumnTransformer` to the housing data by applying the `full_pipeline`'s `fit_transform` method

Note that the `OneHotEncoder` returns a sparse matrix, while the `num_pipeline` will return a dense matrix. When there is a a mix of these two types of matrices, the `ColumnTransformer` estimates the density of the final matrix, and it returns a sparse matrix if the desnsity is lower than a given threshold. By default, this threshold is `sparse_threshold = 0.3`. In this example, it returns a dense matrix. We are ready to go because we have prepared all of my data!

This is not performing great, as most housing prices fall between `$120000` and `$265000`, so a typical error of `$68284` covers this range. Our model is likely underfitting the data, which means that the features do not provide enough information to make good predictions, or that the model is not powerful enough. We can fix underfitting by selecting a more powerful model, to feed the training algorithm with better features, or to reduce the constraints of the model. This model is not regularized, so this rules out the last option. We could try adding more features, but let's try a more complex model to see how it does.

We can try to train a `DecisionTreeRegressor`. This model is capable of capturing complex nonlinear relationships in the data:

````python
from sklearn.tree import DecisionTreeRegressor

# Define the decision tree regression model we are going to use
tree_reg = DecisionTreeRegressor()
# Fit this model to the prepped data and housing labels
tree_reg.fit(housing_prepared, housing_labels)

# Predict the outcomes using the model we just trained
housing_predictions_tree = tree_reg.predict(housing_prepared)
tree_mse = mean_squared_error(housing_labels, housing_predictions_tree)
tree_rmse = np.sqrt(tree_mse)
print("The decision tree's RMSE is:", tree_rmse)
````

Well, this is not surprising, as the decision tree has no regulation on the fitting of this data, so the model is overfitted to the training set.
### Better Evaluation Using Cross-Validation
One way to evaluate the Decision Tree model would be to use the `train_test_split` function to split the training set into a smaller training set and a validation set, then train your models against the smaller training set and evaluate them against the validation set. It will work, but it's a bit more work to do this.

An alternative is to use Scikit-Learn's *cross-validation* feature. The following code performs *K-fold cross validation*. This process randomly splits the training set into 10 distinct (disjoint) subsets called *folds*, then it trains and evaluates the Decision Tree model 10 times, picking out a different fold for evaluation every time and training on the other 9 folds. THe result is an array containing 10 evaluation scores:

````python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring = "neg_mean_squared_error", cv = 10)
tree_rmse_scores = np.sqrt(-scores)
````

Let's look at the results:

````python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean of scores:", scores.mean())
    print("Standard deviation of scores", scores.std())
    
display_scores(tree_rmse_scores)
````

These decision tree models (mostly) did worse than our linear regression model! Notice that cross-validation allows us to get not only an estimate of the performance of your model, but also a measure of how precise this estimate is. The Decision Tree has a score of approximately $70908 \pm 1661$. We do not have this information if you just used one validation set, but cross-validation comes at the cost of training the model several times.

Let's do the same for the linear regression model just to be sure:

````python
lin_scores = cross_val_score(lin_reg, housing_prepared, housing_labels,
                            scoring = "neg_mean_squared_error", cv = 10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
````

This shows that the decision tree is overfitting so badly that it underperforms the linear regression model, as the linear regression's performance is $68481 \pm 1456$. Let's try another model now, the `RandomForestRegressor`, which tains many decision trees on random subsets of the features, then averaging out their predictions. Building a model on top of many other models is known as *ensemble learning*.

````python
from sklearn.ensemble import RandomForestRegressor

forest_reg = RandomForestRegressor()
forest_reg.fit(housing_prepared, housing_labels)
forest_predictions = forest_reg.predict(housing_prepared)
forest_mse = mean_squared_error(housing_labels, forest_predictions)
print("The RMSE of the random forest model is:", np.sqrt(forest_mse))

forest_scores = cross_val_score(forest_reg, housing_prepared, housing_labels,
                               scoring = "neg_mean_squared_error", cv = 10)
forest_rmse_scores = np.sqrt(-forest_scores)
display_scores(forest_rmse_scores)
````

Our random forest performance did pretty well, as our RMSE scores from cross-validation $50316 \pm 1506$ we could do better. We can see that since the RMSE from the random forest model is significantly lower than the validation sets, we are still overfitting the models to the training data. We could do better in this regard by trying other things, such as more complex algorithms or selecting better hyperparameters.
### Fine-Tune Our Model
#### Grid Search
One way to do this fine-tuning is to fiddle with the hyperparameters manually, until we find the right combo of values, but this would be tedious. We can use `GridSearchCV` to search for us. All we need to do is tell if which hyperparams ew want it to experiment with, and what values to try out, and it will evaluate all the possible combinations of hyperparameter values using cross-validation. We try this with the `RandomForestRegressor`:

````python
from sklearn.model_selection import GridSearchCV

param_grid = [
    {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]}
]

forest_reg = RandomForestRegressor()

grid_search = GridSearchCV(forest_reg, param_grid = param_grid, cv = 5,
                          scoring = "neg_mean_squared_error", return_train_score = True)

grid_search.fit(housing_prepared, housing_labels)
````

`GridSearchCV` will try all 12 combinations of hyperparameters in the first dictionary, and then all 6 combinations of hyperparameters in the second dictionary. For all 18 combinations, we perform 5-fold cross validation, so we have 90 training runs for each combination of hyperparameters, which may take a while! We can access the best parameters in a few ways:

````python
# Just call the best parameters
grid_search.best_params_
````

````python
# Call the best estimator
grid_search.best_estimator_
````

````python
# We can access the evaluation scores as well
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]):
    print(np.sqrt(-mean_score), params)
````

### Randomized Search
The grid search approach works when we are exploring only a few combinations, but when the hyperparameter *search space* is large, it is often preferable to use `RandomSearchCV` instead. It works similarly to `GridSearchCV`, but it evaluates a given number of random combinations by selecting a random value for each hyperparameter for each combination.
- This approach makes it easier to explore a larger number of parameter combinations
- It also gives us more control over the computational usage while trying to find the optimal hyperparameters

### Ensemble Methods
We could also fine-tun the model by combining models that perform best, or by putting together an *ensemble*.
